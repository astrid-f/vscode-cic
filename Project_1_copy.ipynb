we can also calculate the mean, median, and mode by using only the Python standard library

before that, we need to re-open the csv file without using Pandas

```python
import csv
data=[] #We can read the csv file with Python Standard Library by making it as a dictionary
with open("NYC_Air_Quality_Data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)
data
```

```python
data=[] #we are doing the same thing as before, reading the csv file as dictionary
with open("NYC_Air_Quality_Data.csv") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        data.append(row)

    total_value = 0
    count = 0
    
    for row in data:
        value = float(row['Data Value'])
        total_value += value #we are accumulating (or summing up) the values of value over multiple iterations of a loop
        count += 1 #here, we are tracking the number of iterations or values that have been processed in the loop
    mean = total_value / count #the average/mean is calculated by the total value divided by count

    print("Mean:", mean)
```

```python
#to calculate the median, we are sorting the list and find the middle value 
# (or the average of the two middle values if there's an even count).
data=[]
with open("NYC_Air_Quality_Data.csv") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        data.append(float(row['Data Value']))
            
sorted_data = sorted(data)
n = len(sorted_data) 

if n % 2 == 1:
    median = sorted_data[n // 2]
else:
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2

print("Median:", median)
```

```python
data=[] #same procedure as before
with open("NYC_Air_Quality_Data.csv") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        data.append(float(row['Data Value']))

# To calculate the mode, we use a dictionary to count how often each value appears.
# Because there can be more than one mode, we store them in a list.
# For each value, we check if its count is higher than the current highest frequency.
# If it is, we update the mode list with this value.

unique_items = set(data) 
max_count = 0 
mode = [] 

for item in unique_items:
    count = data.count(item) 
    if count > max_count: 
        max_count = count 
        mode = [item] 
    elif count == max_count: 
        mode.append(item)

print("Mode:", mode)
```

## Step 4: Data Visualization


```python
import csv

category_sums = {} #We are making an empty dictionary to store the sum of 'Data Value' for each 'Name' category

with open("NYC_Air_Quality_Data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        category = row["Name"]
        value = float(row["Data Value"])

        if category in category_sums: #Adding value to the corresponding category in the dictionary
            category_sums[category] += value
        else:
            category_sums[category] = value

max_length = 55
max_sum = max(category_sums.values())

print("Air Pollutant Prevalence in New York City")
print("-" * 20)

for category, total_value in category_sums.items(): #Learned from Reference 1, shown below
    bar_length = int((total_value / max_sum) * max_length) #Calculate the length of the bar based on the total value
    bar_length = max(bar_length, 1) #Ensure at least 1 symbol, even for very small values
    bar = "༄" * bar_length #Determining the bar length of each category!
    print(f"{category}: {bar} ({total_value:.2f})")
```

## Analysis

The visualization above was made to highlight the key contributors to air quality metrics from our dataset. The goal was to aggregate the total "Data Value" for each category, such as "Carbon monoxide" or "Ozone," and represent these totals visually with proportional bar lengths. This makes it easy to see which categories are the biggest contributors at a glance.

In this visual, "Carbon monoxide" stands out as having the largest impact, with the longest bar, while smaller contributors like "Ozone" are still represented due to a minimum bar length rule I added. This ensures no category is ignored, even if its contribution is relatively small. The research question guiding this was: "Which pollutants are contributing the most to overall air quality levels?" By using proportional bars, I am trying to communicate disparities clearly, helping to prioritize efforts or resources for the largest contributors.
